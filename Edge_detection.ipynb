```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# canny edge detection
import cv2

# Load image
image_canny = cv2.imread('../data/train/alternaria_leaf_spot/000414.jpg', cv2.IMREAD_GRAYSCALE)

# Apply Canny edge detection
edges = cv2.Canny(image_canny, 100, 200)  # Adjust threshold values as needed

# Display results
#cv2.imshow('Original Image', image)
#cv2.imshow('Canny Edge Detection', edges)
#cv2.waitKey(0)
#cv2.destroyAllWindows()

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_canny, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')

plt.show()
```

```python
# sobel edge detection

image = cv2.imread('../data/train/alternaria_leaf_spot/000414.jpg', cv2.IMREAD_GRAYSCALE)

sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = cv2.sqrt(cv2.addWeighted(cv2.pow(sobelx, 2.0), 1.0, cv2.pow(sobely, 2.0), 1.0, 0.0))


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sobel_combined, cmap='gray')
plt.title('Sobel Edge Detection')
plt.axis('off')

plt.show()
```
